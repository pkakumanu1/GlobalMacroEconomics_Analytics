# GDP

```python
GDP = SeriesId["GDP: 81"]
type(GDP)
```

```python
#Joining the list Series Id
StrGDP = GDP.astype(str)
GDPList = '%2C'.join(StrGDP)
GDPList
```

```python
#With passing the list of GDP
url = url = f"https://www.quandl.com/api/v3/datatables/EDIA/ECD?series_id={GDPList}&api_key={api_key}"
print(url)
```

```python
response = requests.get(url).json()
```

```python
col = []
for c in response['datatable']['columns']:
    col.append(c['name'])
```

```python
GDP_df = pd.DataFrame(response['datatable']['data'],columns=col)
```

```python
GDP_df['country'] = GDP_df['description'].apply(lambda  x : x.split(":")[0]  )
GDP_df.tail()
```

```python
GDP_df.to_csv("GDP.csv")
```
