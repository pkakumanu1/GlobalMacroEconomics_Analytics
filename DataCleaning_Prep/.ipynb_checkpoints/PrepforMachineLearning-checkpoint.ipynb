```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
GDP_df = pd.read_csv("GDP.csv")
```

```python
GDP_df.rename(columns = {'value':'GDP'}, inplace = True)
GDP_df
```

```python
UR_df = pd.read_csv("Unemployment_Rate.csv")
```

```python
UR_df.rename(columns = {'value':'UnemploymentRate'}, inplace = True)
UR_df
```

```python
IR_df = pd.read_csv("Inflation_Rate.csv")
```

```python
IR_df.rename(columns = {'value':'InflationRate'}, inplace = True)
IR_df
```

```python
IntR_df = pd.read_csv("InterestRate.csv")
```

```python
IntR_df.rename(columns = {'value':'InterestRate'}, inplace = True)
IntR_df
```

```python
SM_df = pd.read_csv("Stock_Market.csv")
```

```python
SM_df.rename(columns = {'value':'StockMarket'}, inplace = True)
SM_df.dropna()
```

```python
GR_df = pd.read_csv("GoldReserve.csv")
```

```python
GR_df.rename(columns = {'value':'GoldReserve'}, inplace = True)
GR_df
```

```python
COR_df = pd.read_csv("Crude_Oil_Reserve.csv")
```

```python
COR_df.rename(columns = {'value':'CrudeOilReserve'}, inplace = True)
COR_df
```

```python
#result = pd.merge(GDP_df['GDP'], UR_df['UnemploymentRate'], on='country')
result = GDP_df.merge(UR_df, on=['country','date'])[['country','date','GDP','UnemploymentRate']]
```

```python
result = result.merge(IR_df, on=['country','date'])[['country','date','GDP','UnemploymentRate','InflationRate']]
```

```python
result = result.merge(IntR_df, on=['country','date'])[['country','date','GDP','UnemploymentRate','InflationRate','InterestRate']]
```

```python
result = result.merge(SM_df, on=['country','date'])[['country','date','GDP','UnemploymentRate','InflationRate','InterestRate','StockMarket']]
```

```python
result = result.merge(GR_df, on=['country','date'])[['country','date','GDP','UnemploymentRate','InflationRate','InterestRate','GoldReserve']]
```

```python
result = result.merge(COR_df, on=['country','date'])[['country','date','GDP','UnemploymentRate','InflationRate','InterestRate','CrudeOilReserve']]
```
